### Input
| Argument | Type | Default value | Description |
| :--- | :--- | :--- | :--- |
| `chrom_data` | Dict | Required | A dictionary containing the full chromatographic data. The chromatographic data structure, typically generated by MS_Import.jl, which contains the MS1 data ("MS1" key) |
| `n` | Int/Float64 | 5000 - Optional | The number of points to be plotted or for n < 1 it is the percent signal to be plotted. |
| `c_lim` | Float | nothing - Optional | The value for capping the color scale in log10 scale. |
| `interactive`| Bool | false - Optional | Enables an interactive GLMakie window. |
